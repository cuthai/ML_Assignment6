```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('thai_assignment6\output\data.csv')
```

```python
df
```

```python
data = pd.DataFrame(df.loc[(df['Track_Name'] == 'R') &
                           (df['Reset_Type'] == 'Stop_Reset') &
                           (df['Brain_Type'] == 'QLearning')]
                    .groupby('Learning_Rate')['Time'].mean())

# Figure / axis set up
fig, ax = plt.subplots()

# We'll plot the list of params and their accuracy
ax.plot(data['Time'])

# Title
ax.set_title(rf'Learning Curves for R Track, QLearning, using Stop_Reset')

# X axis
ax.set_xlabel('Learning Rate')
ax.set_xlim(0, .2)
ax.set_xticks([.01, .02, .03, .04, .05, .06, .07, .08, .09, .1, .11, .12, .13, .14, .15, .16, .17, .18, .19, .2])
ax.set_xticklabels([.01, .02, .03, .04, .05, .06, .07, .08, .09, .1, .11, .12, .13, .14, .15, .16, .17, .18, .19, .2], rotation=45, fontsize=7)

# Y axis
ax.set_ylabel('Time')
ax.set_ylim(0, 1250)

# Saving
plt.savefig(f'thai_assignment6\output\\R_Track_QLearning_Stop.jpg')
```
